```python
import pandas as pd

vecs_df = pd.read_csv("/home/michal/Documents/misc/thesis/word2bits-in-tensorflow/word2vec_basic.nonbin", 
                      skiprows=[0], header=None, sep=" ")
vecs_df = vecs_df.set_index(0)
# vecs_df = vecs_df.drop(columns=[vecs_df.columns[-1]])
# vecs_df = vecs_df.replace(vecs_df.values.max(), 1)
# vecs_df = vecs_df.replace(vecs_df.values.min(), 0)
# retype to minimal range int to save space
vecs_df = vecs_df.astype('int8')
vecs_df.head(20)
```

```python
from web.embeddings import Embedding
from web.vocabulary import OrderedVocabulary

words = vecs_df.index.values
vectors = vecs_df.values

vocabulary = OrderedVocabulary(words=words)

w = Embedding(vocabulary=vocabulary, vectors=vectors)
```

```python
from web.embeddings import Embedding
from web.vocabulary import OrderedVocabulary

words = set(vecs_df.index.values)
words_vecs = dict()
for w in words:
    try:
        words_vecs[w] = vecs_df.loc[w]
    except KeyError:
        print("Error on %s" % w)
```

```python
vocabulary = OrderedVocabulary(words=list(words_vecs.keys()))

w = Embedding(vocabulary=vocabulary, vectors=list(words_vecs.values()))
```

```python
# Fetch skip-gram trained on GoogleNews corpus and clean it slightly
# w = fetch_SG_GoogleNews(lower=True, clean_words=True)

# Fetch analogy dataset
data = fetch_google_analogy()
```

```python
for cat in (set(data.category)):
    print(cat)
```

```python
# Pick a sample of data and calculate answers
subset = range(0, 18000, 500)
for id in subset:
    w1, w2, w3 = data.X[id][0], data.X[id][1], data.X[id][2]
    print("Question: {} is to {} as {} is to ?".format(w1, w2, w3))
    print("Answer: " + data.y[id])
    try:
        print("Predicted: " + " ".join(w.nearest_neighbors(w[w2] - w[w1] + w[w3], exclude=[w1, w2, w3])))
    except KeyError as e:
        print("some word not in dict")
```
